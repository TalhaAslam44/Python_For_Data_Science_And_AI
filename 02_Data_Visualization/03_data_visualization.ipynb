# Seaborn and Plotting

```python
# data 
df = sns.load_dataset('titanic')
```

```python
# scatter plot

sns.scatterplot(data = df , x= 'age', y='fare', hue='sex')
```

```python
# scatter plot
# scatter plot size of figure
plt.figure(figsize=(12,8))
sns.scatterplot(data = df , x= 'age', y='fare', hue='sex')
```

```python
# scatter plot
# scatter plot size of figure
plt.figure(figsize=(12,8))

# change the scatter point color
palette= {'male': '#056559','female':'blue'}
sns.scatterplot(data = df , x= 'age', y='fare', hue='sex', palette=palette)
```

```python
# scatter plot
# scatter plot size of figure
plt.figure(figsize=(12,8))

# change the scatter point color
palette= {'male': '#056559','female':'blue'}
sns.scatterplot(data = df , x= 'age', y='fare', hue='sex', palette=palette)

# labels
plt.xlabel('age')
plt.ylabel('fare')
plt.title('Scatter plot of age vs fare')
```

```python
# scatter plot
# scatter plot size of figure
plt.figure(figsize=(10,6))

# change the scatter point color
palette= {'male': '#056559','female':'blue'}
sns.scatterplot(data = df , x= 'age', y='fare', hue='sex', palette=palette)

# labels
plt.xlabel('age')
plt.ylabel('fare')
plt.title('Scatter plot of age vs fare')

# saving the plot 300dpi, 600dpi, 1500dpi png
plt.savefig('./plots/scatter_plot.png', dpi=300)
```

```python
# scatter plot
#style of plot
sns.set_theme(style='darkgrid', font_scale=1.5)
# scatter plot size of figure
plt.figure(figsize=(10,6))

# change the scatter point color
palette= {'male': 'blue','female':'red'}
sns.scatterplot(data = df , x= 'age', y='fare', hue='sex', palette=palette)

# labels
plt.xlabel('age')
plt.ylabel('fare')
plt.title('Scatter plot of age vs fare')

# saving the plot 300dpi, 600dpi, 1500dpi png

plt.savefig('./plots/scatter_plot.png', dpi=300)
```

#### Interpertation:

1. We have age of passengers at x-axis and fare at y-axis.
2. The scatter plot shows the relationship between age and fare of titanic passengers.
3. We further grouped the passenger by their genders such as male(blue points) and female(green points).
4. Two Male passengers payed highest fare in titanic, and they might be travleing first class.
5. The oldest person at titanic as also a male having 80 yaers of age as shown in scatter plot.

# Box Plot

```python
df['class'].unique
```

```python
# scatter plot
#style of plot
sns.set_theme(style='darkgrid', font_scale=1.5)
# scatter plot size of figure
plt.figure(figsize=(10,6))

# change the scatter point color
palette= {'First': 'blue','Second':'red', 'Third':'green'}
sns.boxplot(data = df , x= 'sex', y='age', hue='class', palette=palette)

# labels
plt.xlabel('age')
plt.ylabel('fare')
plt.title('box plot of age vs fare')

# saving the plot 300dpi, 600dpi, 1500dpi png

plt.savefig('./plots/box_plot.png', dpi=300)

# change legend position and title
plt.legend(loc='lower right', title='class')
```

```python
# scatter plot
#style of plot
sns.set_theme(style='darkgrid', font_scale=1.5)
# scatter plot size of figure
plt.figure(figsize=(10,6))

# change the scatter point color
palette= {'male': 'blue','female':'red'}
sns.scatterplot(data = df , x= 'age', y='fare', hue='sex',size='age' ,palette=palette)

# labels
plt.xlabel('age')
plt.ylabel('fare')
plt.title('Scatter plot of age vs fare')

# saving the plot 300dpi, 600dpi, 1500dpi png

plt.savefig('./plots/scatter_plot.png', dpi=300)
```

# Boxen Plot


```python
# scatter plot
#style of plot
sns.set_theme(style='darkgrid', font_scale=1.5)
# scatter plot size of figure
plt.figure(figsize=(10,6))

# change the scatter point color
palette= {'First': 'blue','Second':'red', 'Third':'green'}
sns.boxenplot(data = df , x= 'sex', y='age', hue='class', palette=palette)

# labels
plt.xlabel('gender')
plt.ylabel('age')
plt.title('box plot of Age, based on gender and class')

# saving the plot 300dpi, 600dpi, 1500dpi png

plt.savefig('./plots/box_plot.png', dpi=300)

# change legend position and title
plt.legend(loc='lower right', title='class')

# change the c-ticks names
plt.xticks(ticks=[0,1],labels=["Male,","Female"])
```

```python
# scatter plot
#style of plot
sns.set_theme(style='darkgrid', font_scale=1.5)
# scatter plot size of figure
plt.figure(figsize=(10,6))

# change the scatter point color
palette= {'First': 'blue','Second':'red', 'Third':'green'}
sns.barplot(data = df , x= 'sex', y='age', hue='class', palette=palette)

# labels
plt.xlabel('gender')
plt.ylabel('age')
plt.title('box plot of Age, based on gender and class')

# saving the plot 300dpi, 600dpi, 1500dpi png

plt.savefig('./plots/box_plot.png', dpi=300)

# change legend position and title
plt.legend(loc='lower right', title='class')

# change the c-ticks names
plt.xticks(ticks=[0,1],labels=["Male,","Female"])
```
