# Different Types of Plots

```python
# Importing liberaries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the dataset using seaborn   
df = sns.load_dataset('titanic')
df.head()
```

# 1. Scatter Plot

```python
df.plot(kind='scatter', x='age', y='fare')
```

```python
df.plot(kind='scatter', x='age', y='fare', color = 'red')
```

```python
# change colors using hex codes
df.plot(kind='scatter', x='age', y='fare', color = '#FE7F0E')
```

```python
df.plot(kind = 'scatter', x = 'age', y ='fare', color = '#0c822d') 
plt.title('Relationship between Age and Fare')

plt.savefig('./plots/scatter_plot.png', dpi = 300) # for saving plots
plt.show()
```

```python
df.head()
```

# 2. Bar Plot

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
plt.ylabel('Average Fare')
plt.xlabel('Gender')
```

```python
# seaborn plots
# bar plot using seaborn on df
sns.barplot(data = df, x= 'survived', y = 'fare')
```

```python
df.groupby('sex')['age'].mean().plot(kind = 'bar', color = 'red', legend = True)

plt.savefig('./plots/bar_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
df.groupby('sex')['age'].mean().plot(kind = 'bar') # without legend command
plt.show()
```

# 3. Pie Chart

```python
df.head()
```

```python
plt.figure(figsize = (12, 6))
class_dist = df['class'].value_counts()
print(class_dist)
labels = class_dist.index.tolist()
plt.pie(class_dist, labels = labels, autopct = '%1.1f%%', startangle = 180) # The startangle attribute rotates the plot by the specified degrees in counter clockwise direction performed on x-axis of pie chart.
plt.title('The Percentage of passengers in each class')

plt.savefig('./plots/pie_chart', dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
df.head()
```

```python
who_counts = df['who'].value_counts()
print(who_counts)
labels = who_counts.index.tolist()
plt.figure(figsize = (12, 6))
plt.pie(who_counts, labels = labels, autopct = '%1.1f%%')
plt.title('Percentages')

plt.savefig('./plots/pie_chart', dpi = 300, bbox_inches = 'tight')
plt.show()
```

# 4. Histogram

```python
df.head()
```

```python
df['age'].plot(kind = 'hist', bins = 10, edgecolor = 'yellow', color = 'black')
plt.title('Distribution of Age')

plt.savefig('./plots/histogram', dpi = 300, bbox_inches = 'tight')
plt.show()
```

# 5. Line Plot

```python
df.head()
```

```python
age_count = df['age'].value_counts().sort_index()
print(age_count)

plt.figure(figsize = (12, 6))
plt.plot(age_count.index, age_count.values, color = 'black', marker = 'o')
plt.xlabel('Age')
plt.ylabel('Passengers Count')
plt.title('Passengers count by Age')

plt.savefig('./plots/line_plot.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

# 6. HeatMap

# 7. Box Plot

```python
df.head()
```

```python
plt.figure(figsize = (12, 6))
df[['age']].boxplot(grid = False, vert = True)
plt.title('Outliers in Age')

plt.savefig('./plots/box_plot_using_pandas.png', dpi = 300)
plt.show()
```

```python
# Similarly, we can plot box-plot using seaborn liberary
plt.figure(figsize = (12, 6))
sns.boxplot(df[['age']])
plt.title('Outliers in Age')

plt.savefig('./plots/box_plot_using_seaborn.png', dpi = 300)
plt.show()
```

# 8. Area Chart

```python
df.head()
```

```python
# Creating a new column to categorize passengers into age groups
age_bins = [0, 18, 30, 50, 100]  # Define age group bins
age_labels = ['0-17', '18-29', '30-49', '50+']  # Define labels for age groups

df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

# Calculating the cumulative count of passengers by age group
passenger_counts = df.groupby('age_group').size().cumsum()
plt.figure(figsize = (12, 6))
# Creating the area plot
plt.fill_between(passenger_counts.index, passenger_counts.values, alpha=0.6) # aplha shows the transparency

# Customizing the area plot
plt.xlabel('Age Group')
plt.ylabel('Cumulative Passengers')
plt.title('Cumulative Passengers by Age Group')
plt.xticks(rotation=45)

# Saving and Displaying the area plot
plt.savefig('./plots/area_chart.png', dpi = 300)
plt.show()
```
